```python
# Author: Matthew Jacobsen
# Date Created: 08/23/19
# Date Last Modified: 10/22/2019
# Purpose: Generate Fox advancement chart based on Google Sheet Data

# Step 1: acquire data automatically from Trail Life Master Google Sheet
# stub: def googleSheetConnector():

# Main function
def mainChartFunction():
    chartSetup()
    heritage()
    
def chartSetup():
    
    # Create file object to write to
    foxChart = open('foxadvancement.html', 'w')

    # Establish HTML document header and styles
    messageHeader = """<!DOCTYPE html>
        <html>
            <head>
                <style>
                    table, td, tr {
                        border: 1px solid black; 
                        border-collapse: collapse;
                    }
                    .tr1 {color: white;
                        border: 1px solid black; 
                        border-collapse: collapse;
                    }
                    th {width: 125px;
                        height: 125px;}
                    .tdCoreStep
                        {background-color: gray}
                    .tdTGI
                        {width:100px;}
                </style>
            </head>"""

    # Create table to hold advancement contents
    tableHeader = """
        <body>
            <table style="width:100%">
                <tr class="tr1">
                    <th colspan="2">
                        <img src="https://res.cloudinary.com/imaginatewithme/image/upload/v1566608229/foxes.png" width="100" height="100">
                    </th>
                    <th bgcolor="#806b34">Heritage</th>
                    <th bgcolor="black">Hobbies</th>
                    <th bgcolor="#731f42">Life Skills</th>
                    <th bgcolor="#1e5f8a">Outdoor Skills</th>
                    <th bgcolor="#f7a923">Science & Technology</th>
                    <th bgcolor="#64b053">Sports & Fitness</th>
                    <th bgcolor="#de0430">Values</th>
                </tr>
                <tr>
        """

    # Create row for Heritage Branch
def heritage():
    heritageRowStart = """
        <td class="tdCoreStep">Core Step</td>
            <td>
                <table style="width:100%">
                    <tr>
                        <td>Date</td>
                    </tr>
                    <tr>
                        <td>Step #</td>
                    </tr>
                    <tr>	
                        <td class="tdTGI">Trail Guide Initials</td>
                    </tr>
                </table>
            </td>
        """
    rowData = ""
    counter = 0
    while counter <= 6:
        Date = "1/1/111{}".format(counter)
        Step = 2+counter
        TGI = "MJ"
        rowData += """
        <td>
            <table style="width:100%">
                <tr>
                    <td>{Date}</td>
                </tr>
                <tr>
                    <td>{Step}</td>
                </tr>
                <tr>
                    <td>{TGI}</td>
                </tr>
            </table>
        </td>
        """.format(Date=Date, Step=Step, TGI=TGI)
        counter += 1
        foxChart.write(messageHeader+tableHeader+rowStart+rowData)
    foxChart.close()
```
