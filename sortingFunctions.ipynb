```python
import random

list = []

def getRand():
    newNum = random.randint(1,10)
    return newNum

i = 0
while i < 10:
    list.append(getRand())
    i+=1
```

```python
i = 0
while i < len(list):
    print(i, " ", list[i])
    i+=1
```

```python
def bubbleSort():
    x = 0
    counter = 0
    while x < len(list): 
        i = 0
        while i < len(list)-1:
            firstPos = list[i]
            secondPos = list[i+1]
            print("evaluating ", firstPos, "and ", secondPos)
            if firstPos > secondPos:
                print(firstPos, " is greater than ", secondPos)
                list[i] = secondPos
                list[i+1] = firstPos
                print(secondPos, " is now switched with ", firstPos)
                print(list)
            i+=1
            counter+=1
        x +=1
        counter+=1
        print(list)
    print(counter)
    
```
