```python
'''
MarkLogic REST Example
References:
    https://docs.marklogic.com/guide/rest-dev/documents#id_11953
'''

import requests
import getpass
from requests.auth import HTTPBasicAuth

# Url of the MarkLogic Server
# MARKLOGIC_SERVER = 'http://rx-marklogic1.rx.afrl-wrs.hpc.mil:8032/'
MARKLOGIC_SERVER = 'http://localhost:3003/'
MARKLOGIC_ENDPOINT = 'LATEST/documents/'
MARKLOGIC_SEARCH = 'v1/search/'
MARKLOGIC_USERNAME = getpass.getpass('Username for MarkLogic: ')
MARKLOGIC_PASSWORD = getpass.getpass('Password for MarkLogic: ')

```

```python
# READ FILE HERE

# Document to store into MarkLogic
document = {
    'Project': 'Steel for ',
    'key4': 'value2',
    'program_elements': [
        '123456', '13579'
    ]
}

# Create a document URI (This acts as a key to make the whole document unique)
#  Suggestion: Use UUID4

params = {
    'uri': '/test/test_document'
}

# Use a PUT method to store the data into MarkLogic
url = MARKLOGIC_SERVER + MARKLOGIC_ENDPOINT
r = requests.put(
    url=url, 
    params=params, 
    json=document, 
    verify=False, 
    auth=HTTPBasicAuth(MARKLOGIC_USERNAME, MARKLOGIC_PASSWORD)
)

print(r.headers)
print(r.content)
```

```python
print(r.status_code)
```

```python
# Now let's get the document back
url = MARKLOGIC_SERVER + MARKLOGIC_SEARCH

params = {
    'collection': 'granta.record.NG6.Processing Pedigree',
    'options': 'applayer',
    'format': 'json',
    'q': 'results:raw'
}

r = requests.get(
    url=url, 
    params=params,      
    verify=False, 
    auth=HTTPBasicAuth(MARKLOGIC_USERNAME, MARKLOGIC_PASSWORD)
)

# print('Headers {}'.format(r.headers))
#print('Content: {}'.format(r.content))
json_data = r.json()
len(json_data)
```
